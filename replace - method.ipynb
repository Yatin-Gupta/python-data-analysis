```python
import pandas as pd
import numpy as np

df = pd.read_csv("./practice1.csv")
df
```

```python
df.replace([-9999,-8888], np.NaN)
```

```python
# Replace column wise
df.replace({
    'temperature':[-9999, -8888],
    'windspeed':-9999,
    'event':'No Event'
}, np.NaN)
```

```python
# Replace by values
df.replace({
    -9999:-7777,
    -8888:np.NaN,
    'No Event':0
})
```

```python
# Replace all alphabets
df.replace("[A-Za-z]","",regex=True)
```

```python
df.replace({
    "temperature":"[A-Za-z]",
    "windspeed":"[A-Za-z]"
},"",regex=True)
```
